```python
f = open("EIGRP_in_1.txt", "r")

class Switch:
    def __init__(self, ID):
        self.ID = ID
        self.Path = {}
    
    def addPath(self, targetID, nextHop, cost):
        self.Path[targetID] = {"nextHop" : nextHop, "cost" : cost}
    
    def modRoot(self, target, bandwidth):
        self.RootID = target.RootID
        self.RootPriority = target.RootPriority
        self.RootPortTo = target.ID
        self.Cost = target.Cost + self.bandwidth_to_cost_table[bandwidth]
    
    def getInfo(self):
        s = "ID:{}".format(self.ID)
        s += "\npaths:"
        for dest in sorted(self.Path.keys()):
            s += "\n{}:\n - nextHop:{}\n - cost:{}".format(dest, self.Path[dest]["nextHop"], self.Path[dest]["cost"])
        return s
    
    def old_getInfo(self):
        s = "ID:{}".format(self.ID)
        s += "\npaths:"
        for dest in self.Path:
            s += "\n{}:\n - nextHop:{}\n - cost:{}".format(dest, self.Path[dest]["nextHop"], self.Path[dest]["cost"])
        return s
    
myID = int(f.readline())
n = int(f.readline())
mainSwitch = Switch(myID)
for i in range(n):
    inp = f.readline().split()
    mainSwitch.addPath(targetID = int(inp[0]), nextHop = int(inp[0]), cost = int(inp[1]))
print("Hello")
# print(mainSwitch.getInfo())

m = int(f.readline())
for i in range(m):
    recv_from_ID = int(f.readline())
    s = int(f.readline())
    targetDestinations = []
    needUpdate = False
    needHello = False
    for i in range(s):
        line = f.readline().split()
        destinationID = int(line[0])
        targetDestinations.append(destinationID)
        distance = int(line[1]) + mainSwitch.Path[recv_from_ID]["cost"]
        if not destinationID in mainSwitch.Path:
            mainSwitch.addPath(targetID = destinationID, nextHop = recv_from_ID, cost = distance)
            needHello = True
        elif mainSwitch.Path[destinationID]["cost"] > distance:
            mainSwitch.Path[destinationID]["nextHop"] = recv_from_ID
            mainSwitch.Path[destinationID]["cost"] = distance
            needHello = True
        else:
            needUpdate = True
    for myDest in mainSwitch.Path:
        if not myDest in targetDestinations:
            needUpdate = True
    
    if needUpdate and needHello:
        print("Update Hello")
    elif needUpdate:
        print("Update")
    elif needHello:
        print("Hello")

print(mainSwitch.getInfo())
```
